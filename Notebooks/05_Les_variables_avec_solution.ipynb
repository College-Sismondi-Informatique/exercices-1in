### Apporter des variations √† des r√©p√©titions

On peut modifier une variable √† chaque it√©ration d'une boucle `for`. On peut, par exemple, utiliser une variable pour afficher une table de mutiliplication. V√©rifier que le programme suivant affiche bien la table de multiplication de 7:

````python
n = 1
for _ in range(10):
    print(n, " x 7 =", n*7)
    n = n + 1
````

Pour ajouter 1 √† la variable `n`, on √©crit `n = n + 1` qui consiste √† remplacer la valeur de `n` par son ancienne valeur plus 1.

On peut utiliser un proc√©der similaire pour tracer une spirales, comme le fait le programme suivant:

````python
from turtle import *
        
x = 20
for _ in range(30):
    forward(x)
    left(90)
    x = x + 10
    
done()
````

### Variables avec une fonction avec param√®tre. 

On peut utiliser la valeur d'une variable comme argument d'une fonction. Le programme suivant va permettre de dessiner des triangles de plus en plus grands. En observant le code, vous constatez que la fonction ```triangle``` a ```longueur``` comme param√®tre (ligne 3) et ```x``` comme argument (ligne 10). Ce qui signifie que ```longueur``` prendra la valeur de ```x```.

````python
from turtle import *

def triangle(longueur):
    for _ in range(3):
        forward(longueur)
        left(120)
        
x = 20
for _ in range(4):
    triangle(x)
    forward(x + 10)
    x = x + 20
    
done()
````

## L'indice d'une boucle `for`, une variable gratuite

Il est possible de remplacer le symbole `_` dans une boucle `for` par le nom d‚Äôune variable, appel√©e indice de boucle. Cette variable est initialis√©e √† `0`, puis incr√©ment√©e de `1` √† chaque it√©ration de la boucle. Par exemple, le programme suivant :
 
```python
k = 0
for _ in range(10):
    print(k)
    k = k + 1
```

est √©quivalent au programme suivant:
```python
for k in range(10):
    print(k)
```
