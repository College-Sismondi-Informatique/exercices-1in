### Docstring
Chaque fois qu’une fonction est définie, il est demandé d’ajouter un commentaire de type **docstring** (c’est un commentaire encadré par des triple apostrophes `''' ... '''`) qui indique :
- ce que fait la fonction,
- si elle a des paramètres,
- quels sont leurs types,
- et ce que représentent ces paramètres.

**Exemple :**


```python
def triangle(lg_cote):
    ''' dessine un triangle équilatéral
    param(s) :
    lg_cote (int) : longueur d'un côté du triangle

    Exemples
    '''
    for i in range(3):
        forward(lg_cote)
        right(120)
```
