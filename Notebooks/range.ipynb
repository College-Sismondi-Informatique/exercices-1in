### Les différents paramètres de la fonction `range`

!!! info 1 paramètre
On a jusqu'ici utilisé la fonction `range` qu'avec un seul paramètre, le nombre de répétitions :

```python
for k in range(10):
    print(k)
```

Ici `10` indique que l'on veut s'arrêter juste avant que `k` vaille 10. `k` va donc aller de 0 à 9. 
!!!
!!! info 2 paramètres
Mais comment faire pour aller de 10 à 20 ? On peut utiliser 2 paramètres de `range`. Dans ce cas, le 1er spécifie la valeur de laquelle on part.

```python
for k in range(10, 20):
    print(k)
```

!!! info 3 paramètres

Et comment afficher uniquement les nombres pairs ?  On peut utiliser 3 paramètres de `range`. Dans ce cas, le 3ème spécifie la quantité de laquelle on augmente à chaque tour. Par exemple, le programme suivant va afficher les nombres **pairs** de 10 à 19.
```python
for k in range(10, 20, 2):
    print(k)
```
